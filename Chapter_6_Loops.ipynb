**Exercise.1:**

Write a loop that prompts the user to enter a series of pizza toppings until they enter a 'quit' value. As they enter each topping,

print a message saying you’ll add that topping to their pizza.

```python
#loop that prompts that user enters for the pizza toppings they like

pizza_toppings='\nEnter the name of toppings you want:'   #asking user to print the name of toppings
pizza_toppings+='\nEnter quit when you are done'  
while True:
  pizza=input(pizza_toppings)
  
  if pizza=='quit':
    break
  else:
    print('I want to have',pizza.title(),'!')   
```

Exercise.4:

Make a list called sandwich_orders and fill it with the names of various sandwiches. Then make an empty list called finished_sandwiches. Loop through the list of sandwich  orders and print a message for each order, such as I made your tuna sandwich. As each sandwich is made, move it to the list of finished sandwiches. After all the sandwiches have been made, print a message listing each sandwich that was made.

```python
#a program to make sandwiches order with a message
#list of sandwich orders with filling name of various sandwichs

sandwich_orders = ['cheese','beef','chicken','vegetable']
finished_sandwich = []

while sandwich_orders:
  current_sandwich = sandwich_orders.pop()
  print("Enjoy your" + current_sandwich +"sandwich.")
  finished_sandwich.append(current_sandwich)

print("\n")
print('Sandwich Orders:')
for sandwich in finished_sandwich:
  print(sandwich,'Sandwich')
```
